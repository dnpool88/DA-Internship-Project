```python
def checkIfXPATH(driver, string):
    """ Check if element exists """
    try:
        path = driver.find_element(By.XPATH, string)
    except:
        return False
    return path
```

```python
def checkIfCLASS(driver, string):
    """ Check if element exists """
    try:
        path = driver.find_element(By.CLASS_NAME, string)
    except:
        return False
    return path
```

```python
def checkIfID(driver, string):
    """ Check if element exists """
    try:
        path = driver.find_element(By.ID, string)
    except:
        return False
    return path
```

```python
def waitPath_ID(driver, string):
    """ Wait for path to become available """
    check = False
    count = 0
    while check is False:
        path = checkIfID(driver, string)
        if path is False:
            time.sleep(.5)
        else:
            check = True
        count += 1
        if count > 20:
            check = True
    check = False
    
```

```python
def waitPath_CLASS(driver, string):
    """ Wait for path to become available """
    check = False
    count = 0
    while check is False:
        path = checkIfCLASS(driver, string)
        if path is False:
            time.sleep(.5)
        else:
            check = True
        count += 1
        if count > 20:
            check = True
    check = False
```

```python
def waitPath_XPATH(driver, string):
    """ Wait for path to become available """
    check = False
    count = 0
    while check is False:
        path = checkIfXPATH(driver, string)
        if path is False:
            time.sleep(.5)
        else:
            check = True
        count += 1
        if count > 20:
            check = True
    check = False
```

```python
def fastPath_XPATH(driver, string):
    """ Wait for path to become available """
    check = False
    count = 0
    while check is False:
        path = checkIfXPATH(driver, string)
        if path is False:
            time.sleep(.5)
        else:
            check = True
        count += 1
        if count > 10:
            check = True
    check = False
```

```python
def scrape_data(driver, specific_type):
    main_df = pd.DataFrame()
    count = 1
    
    # Loop through pages
    while count < 5:
        linkedIn_count = 0
        websites = []
        linkedIns = []
        
        # Read table into dataframe
        page_source = driver.page_source
        companyInfo = (pd.read_html(page_source))[0]
        comp_df = companyInfo.drop(columns = ['Actions', 'Industry'])
        
        # Check if at the end of all pages
        waitPath_XPATH(driver, '//*[@id="provider-mounter"]/div/div[2]/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div/div/div/div/div[3]/div/div/div/div/div[3]/div/div[2]/div/div/div/div[1]/div[1]')
        pg = driver.find_element(By.XPATH, '//*[@id="provider-mounter"]/div/div[2]/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div/div/div/div/div[3]/div/div/div/div/div[3]/div/div[2]/div/div/div/div[1]/div[1]').text
        print('Page #: ' + str(pg))
        curr_pg = int(pg)
        if count == 1:
            past_pg = curr_pg
        else:
            if past_pg == curr_pg:
                count = 5
                continue
            else:
                past_pg = curr_pg
        
        # Create temporarily dataframe and save to company df
        comp_df = comp_df[comp_df["Company"].str.contains("Enrichment Available") == False]
        companies_count = len(comp_df.index)
        
        # Create  empty lists of websites and linkedIns
        while len(websites) < companies_count:
            websites.append(None)
            linkedIns.append(None)
        
        # Get all links on the page
        soup = BeautifulSoup(page_source, 'html.parser')
        links = soup.find_all('a', class_='zp-link zp_3_fnL')

        for link in links:
            line = link['href']
            
            if 'http' not in line:
                continue
            
            if 'twitter' in line:
                continue
            elif 'facebook' in line:
                continue
            elif 'linkedin' in line:
                linkedIns[linkedIn_count] = line
                linkedIn_count += 1
            else:
                websites[linkedIn_count] = line
                
        # Add lists to dataframe
        comp_df['Company Websites'] = websites
        comp_df['LinkedIn'] = linkedIns
        
        # Run method to find executives and add to a dataframe
        contact_df, revenue = navigate_comp(driver)
        
        # Add revenue and specialty to dataframe
        comp_df['Revenue'] = revenue
        
        type_ = [specific_type] * len(comp_df.index)
        comp_df['Type'] = type_
        
        temp_df = pd.concat([comp_df, contact_df], axis=1)
        main_df = pd.concat([main_df, temp_df])
        
        # Get Next Page element
        next_elm = driver.find_elements(By.CLASS_NAME , 'zp_2cHnw')
        next_page = next_elm[-1]
        
        # Try to click next page
        try:
            next_page.click()
        except WebDriverException:
            print('Next Page Problem!')
            count = 5
            continue
        
        time.sleep(5)
        count += 1
        
    # Export data to a csv file   
    print('Total companies exported: ' + str(len(main_df.index)))
    with open('Apollo_Data.csv', 'a') as f:
        f.write('\n') 
    path = ('/Users/dylanpool/KVLR Capital/Final_Apollo_Data.csv')
    #main_df.to_csv(path, mode='w', index=False)
    main_df.to_csv(path, mode='a', header=False, index=False)
        
```
